```python
"""
Sequential model com Kera
"""
# importando biblitecas
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import accuracy_score

# constantes
PATH_FILE = 'data/diabetes.csv'
RANDOM_STATE = 42
TEST_SIZE = .2

# fixando seed para reproducao
np.random.seed(RANDOM_STATE)

# leitura dos dados
df_diabetes = pd.read_csv(PATH_FILE)

# separando entre labels e targets
X = df_diabetes.drop(columns='Outcome')
y = df_diabetes.loc[:, 'Outcome']

# separando entre dados de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE)

# criando o modelo
model = Sequential()
model.add(Dense(32, input_dim=8, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# treinando o modelo
model.fit(X_train, y_train, epochs=1_000, batch_size=10)

# avaliando modelo
score = model.evaluate(X_test, y_test)
y_pred_proba = model.predict(X_test)
```

```python
print(f'Acc keras: {score[1]*100:.2f}%')
```

```python
from sklearn.ensemble import RandomForestClassifier

rand_forest = RandomForestClassifier()
rand_forest.fit(X_train, y_train)

y_pred = rand_forest.predict(X_test)
print(f'Acc teste Rand Forest: {accuracy_score(y_test, y_pred)*100:.2f}%')
```
